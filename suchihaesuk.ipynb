```python
class NumSet(set):
    def __init__(self, data, f_tol=1e-7):
        self.f_tol = f_tol
        s = set()
        for d in sorted(data, key=lambda e:len(str(float(e)))):
            d = float(d)
            for ele in s:
                if abs(ele-d) <= f_tol:
                    break
            else:
                s.add(d)
        super().__init__(s)
```

## 방정식의 수치적 해석

이럴 경우 수치 해석법으로 접근할 수 있다.

파이썬에는 scipy.optimize라는 모듈을 범용 다차원 문제 해결에 이용할 수 있다.

일반적인 비선형 문제는 broyden1(), broyden2() 함수를 이용할 수 있다

```python
# ipython notebook restart 한 후 실행할 것
```

```python
%pylab inline
```

```python
def f(x):
   return x - 2 * sin(x)
```

```python
# 대락적인 근의 범위를 그래프로 확인하기

x = linspace(-10, 10)
plot(x, f(x))
grid()
```

```python
# -5~5 사이에서 근 구하기
import scipy.optimize

roots = [scipy.optimize.broyden2(f, xin, f_tol=1e-14) for xin in arange(-5, 5)]
roots
```

```python
# 오차 범위 내의 해를 정리하려면 앞서 정의한 NumSet 클래스를 활용할 수 있다.
NumSet(roots, f_tol=1e-9)
```

### Kepler 방정식의 해 구하기

$f(E) = M - E + esin(E) = 0$

sympy의 해는 없음

```python
from functools import partial

def f(E, M, e):
    return M -  E + e*sin(E)

f1 = partial(f, M=0.5, e=0.2)

scipy.optimize.broyden2(f1, 1, f_tol=1e-14)
```

```python
# 검산해보기
f1(0.6154681694899552)
```

```python
x = linspace(-5, 5)
plot(x, f1(x))
grid()
```

```python
# M=0.7, e=0.3일 때의 Kepler 방정식
f2 = partial(f, M=0.7, e=0.2)
scipy.optimize.broyden2(f2, 1, f_tol=1e-14)
```

### $ 2x - tan(x) = 0$

```python
def f(x):
    return 2*x - np.tan(x)
```

```python
x = linspace(-10, 10, 300)
ylim(-20, 20)
plot(x, f(x))
grid()
```

```python
import scipy.optimize

scipy.optimize.broyden2(f, -5, f_tol=1e-14) 
```

```python
roots = []
for xin in arange(-6, 6):
    try:
        r = scipy.optimize.broyden2(f, xin, f_tol=1e-14) 
        roots.append(r)
    except:
        continue
```

```python
NumSet(roots)
```

# 선형방정식 계의 해

$\left\{\begin{matrix} 3x+y=9 & \mbox{ } \\ x+2y=8 & \mbox{ } \end{matrix}\right.$

```python
a = np.array([[3,1], [1,2]])
b = np.array([9,8])

x = np.linalg.solve(a, b)
x
```

```python
# 검산
np.allclose(np.dot(a, x), b)
```

$\left\{\begin{matrix} 
    0.24x+0.36y+0.12z=0.852 & \mbox{ } \\ 
    0.12x+0.16y+0.24z=0.620 & \mbox{ } \\
    0.15x+0.21y+0.25z=0.740
         \end{matrix}\right.$

```python
a = np.array([[0.24, 0.36, 0.12], 
              [0.12, 0.16, 0.24],
              [0.15, 0.21, 0.25]])
b = np.array([0.852, 0.620, 0.740])

x = np.linalg.solve(a, b)
x
```

```python
# 검산
np.allclose(np.dot(a, x), b)
```

### Interpolating N-Dimensional data

```python
from scipy.interpolate import griddata

def func(x, y):
    return x*(1-x)*np.cos(4*np.pi*x) * np.sin(4*np.pi*y**2)**2
```

```python
# points 좌표 값, values 값
points = np.random.rand(1000, 2)
values = func(points[:,0], points[:,1])
```

```python
# 알고 싶은 좌표의 값
grid_x, grid_y = np.mgrid[0:1:100j, 0:1:200j]
```

```python
grid_z0 = griddata(points, values, (grid_x, grid_y), method='nearest')
grid_z1 = griddata(points, values, (grid_x, grid_y), method='linear')
grid_z2 = griddata(points, values, (grid_x, grid_y), method='cubic')
```

```python
imshow(grid_z0.T, extent=(0,1,0,1), origin='lower')
```

```python
imshow(grid_z1.T, extent=(0,1,0,1), origin='lower')
```

```python
imshow(grid_z2.T, extent=(0,1,0,1), origin='lower')
```

### 1D Spline interpolation


```python
from scipy.interpolate import splrep, splev

x = np.arange(0, 2*np.pi+np.pi/4, 2*np.pi/8)
y = np.sin(x)
tck = splrep(x, y, s=0) # 값 설정

# 보간
xnew = np.arange(0,2*np.pi,np.pi/50)
ynew = splev(xnew, tck, der=0)
```

```python
plot(x, y, xnew, ynew)
```

### 2D Spline interpolation

```python
# 2D 값 설정
x, y = np.mgrid[-1:1:20j, -1:1:20j]
z = (x+y) * np.exp(-6.0*(x*x+y*y))
```

```python
pcolor(x, y, z)
```

```python
# 보간 하기
from scipy.interpolate import bisplrep, bisplev

xnew, ynew = np.mgrid[-1:1:70j, -1:1:70j]
tck = bisplrep(x, y, z, s=0)
znew = bisplev(xnew[:,0], ynew[0,:], tck)
```

```python
pcolor(xnew, ynew, znew)
```

### Using radial basis functions for smoothing/interpolation

http://docs.scipy.org/doc/scipy-0.14.0/reference/tutorial/interpolate.html

This example compares the usage of the Rbf and UnivariateSpline classes from the scipy.interpolate module.

```python
from scipy.interpolate import Rbf, InterpolatedUnivariateSpline
```

```python
# setup data
x = np.linspace(0, 10, 9)
y = np.sin(x)
```

```python
# use fitpack2 method
ius = InterpolatedUnivariateSpline(x, y)
xi = np.linspace(0, 10, 101)
yi = ius(xi)
```

```python
plot(x, y, 'o')
plot(xi, yi)
```

### 2D

```python
# 2-d tests - setup scattered data
x = np.random.rand(100)*4.0-2.0
y = np.random.rand(100)*4.0-2.0
z = x*np.exp(-x**2-y**2)
```

```python
# use RBF
ti = np.linspace(-2.0, 2.0, 100)
XI, YI = np.meshgrid(ti, ti)

rbf = Rbf(x, y, z, epsilon=2)
ZI = rbf(XI, YI)
```

```python
pcolor(XI, YI, ZI)
```
