# Pickle

## Pickle to file

```python
 import pickle

t = (1,2,3, 'python', 'rules')  # tuple
s = {1,2,3,4,5}  # set
l = [3,4,5] # list
d = {'tuple':t, 'set':s, 'list':l}

with open('test1.pickle', 'wb') as f:
    pickle.dump(t, f)
    pickle.dump(s, f)
    pickle.dump(d, f)
```

```python
with open('test1.pickle', 'rb') as f:
    print (pickle.load(f))
    print (pickle.load(f))
    print (pickle.load(f))
```

## Pickle to string

```python
s = pickle.dumps(t)
```

```python
s
```

```python
pickle.loads(s)
```

## 노드 탐색

 * findall() - 하위 노드(children) 중 특정 노드 검색
 * find() - 첫 번째 엘리먼트
 * iter() - 모든 하위 노드에서 검색

```python
# 모든 wpt 찾기
for wpt in root.findall('{http://www.topografix.com/GPX/1/1}wpt'):
    lat = wpt.get('lat')
    lon = wpt.get('lon')
    name_ele = wpt.find('{http://www.topografix.com/GPX/1/1}name')
    print (name_ele.text, lat, lon)
```

```python
# 모든 trkpt 찾기 ???
root.findall('{http://www.topografix.com/GPX/1/1}trkpt')
```

```python
root[6].getchildren()
```

```python
root[6][1].findall('{http://www.topografix.com/GPX/1/1}trkpt') [:10]
```

```python
# 모든 trkpt 찾기1 - iter() !!!
root.iter('{http://www.topografix.com/GPX/1/1}trkpt')
```

```python
for trkpt in root.iter('{http://www.topografix.com/GPX/1/1}trkpt'):
    lat, lon = trkpt.get('lat'), trkpt.get('lon')
    ele = trkpt.find('{http://www.topografix.com/GPX/1/1}ele').text
    tm = trkpt.find('{http://www.topografix.com/GPX/1/1}time').text
    print (tm, lat, lon, ele)
```

```python
# 모든 trkpt 찾기2 - findall() !!!
ns = '{http://www.topografix.com/GPX/1/1}'
root.findall('./{0}trk/{0}trkseg/{0}trkpt'.format(ns))
```

```python
# 모든 trkpt 찾기3 - findall() !!!
ns = '{http://www.topografix.com/GPX/1/1}'
root.findall('.//{0}trkpt'.format(ns))
```

```python
# 모든 trkpt 찾기3 - findall() !!!
ns = '{http://www.topografix.com/GPX/1/1}'
root.findall('.//{0}trkpt[1]'.format(ns))   # 첫 번째 노드만 선택
```

```python
root.findall('.//{0}trkpt[last()]'.format(ns))   # 마지막 노드만 선택
```

```python
# 모든 노드들 중에서 'lat' 속성이 있는 노드들만 탐색
root.findall('.//*[@lat]')
```

# Excel file

```python
import pandas as pd
import numpy as np

wb = pd.ExcelFile('G:\\141201\\simple.xls')

print (wb.sheet_names)
df = wb.parse("Sheet 3")
print (df.head())
```

```python
X = list(df.index)
Y = list(df.columns)
```

```python
X2, Y2 = np.meshgrid(X, Y)
```

```python
Z = df.values
```

```python
%pylab inline
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import matplotlib.pyplot as plt
from matplotlib.ticker import LinearLocator, FormatStrFormatter

fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X2, Y2, Z, rstride=1, cstride=1, cmap=cm.coolwarm,
        linewidth=0, antialiased=False)

ax.set_zlim(-1.01, 1.01)

ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
