```python
d=1
deg =0.0
for num, den in gps_lat:
    deg+=(num/den)/d
    d*=60
deg
```

```python
def gps2deg(gps):
    d=1
    deg =0.0
    for num, den in gps:
        deg+=(num/den)/d
        d*=60
    return deg
```

```python
lat, lon= gps2deg(gps_lat),gps2deg(gps_lon)
print(lat, lon)
```
