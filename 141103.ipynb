```python
##최대공약수
def gcd(p, q):
    if (q == 0):
        return p
    return gcd(q, p%q)
```

```python
gcd(12,8)
```

```python
def disjoint(a,b):
    return gcd(a,b)==1
```

```python
L=[k for k in range(2,3120) if disjoint(3120,k)]
```

```python
def pmod(c,d,N):
    r=1
    for i in range(d):
        r=r*c%N
    return r
```

```python
def isprime(n):
    for i in range(2,math.floor(n**0.5)+1):
        if n%i==0:
            return False
    return True
```
